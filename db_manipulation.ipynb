```python
import pandas as pd
import pymysql, os
from sqlalchemy import create_engine
pymysql.install_as_MySQLdb()
```

```python
fred_df = pd.read_csv("data/correct_fred_data.csv")
```

```python
fred_df.head(10)
```

```python
fred_df.shape
```

```python
PASSWD = os.getenv("LOCAL_MYSQL_PASS")
USER = "root"
PORT = "3306"
DOMAIN = "localhost"
DB = "fred_db"
engine = create_engine(f"mysql://{USER}:{PASSWD}@{DOMAIN}:{PORT}/{DB}")
conn = engine.connect()
```

```python
fred_df.to_sql(name='fred_data', con=engine, if_exists='replace', index=False)
```

```python
df = pd.read_sql_query('select * from fred_data', con=engine)
```
